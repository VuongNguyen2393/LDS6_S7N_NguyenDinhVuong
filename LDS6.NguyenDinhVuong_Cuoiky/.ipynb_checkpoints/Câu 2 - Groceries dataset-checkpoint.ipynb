# Cho dữ liệu ItemList.xlsx
Yêu cầu: Đọc dữ liệu, chuẩn hóa dữ liệu (nếu cần) và **chọn một thuật toán thích hợp trong nhóm Association rule learning** để tính toán mức độ kết hợp giữa các item.
1. Áp dụng thuật toán (tự lựa chọn các tham số phù hợp cho thuật toán, lưu ý với số lượng transaction càng nhiều thì các ngưỡng nên đặt càng nhỏ). In kết quả. Vẽ biểu đồ.
2. Tìm kiếm thông tin từ kết quả: trong thông tin kết quả có 'sausage' không? Nếu có thì 'sausage' kết hợp với item nào?
3. Cho biết 15 sản phẩm được mua nhiều nhất. Vẽ biểu đồ.
4. Cho biết 15 sản phẩm được mua ít nhất. Vẽ biểu đồ.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
data = pd.read_excel('./du lieu cung cap/ItemList.xlsx',header=None)
```

```python
data.head()
```

```python
data.shape
```

```python
item_list = []
for i in range(0,data.shape[0]):
    row = []
    for j in range(0,data.shape[1]):
        row.append(str(data.iloc[i,j]))
    item_list.append(row)
```

```python
#Chuyển dữ liệu về dạng chuẩn trước khi áp dụng thuật toán:
te = TransactionEncoder()
te_array = te.fit_transform(item_list)
df = pd.DataFrame(te_array,columns=te.columns_)
```

```python
# Xóa xột nan khỏi dữ liệu:
df.drop('nan',axis=1,inplace=True)
df.head()
```

##### Nhận xét: 
Sau khi biến đổi dữ liệu về dạng chuẩn, xét thấy số lượng dữ liệu lớn nên áp dụng thuật toán Apriori để tính toán độ kết hợp của các items.

### Áp dụng thuật toán (tự lựa chọn các tham số phù hợp cho thuật toán, lưu ý với số lượng transaction càng nhiều thì các ngưỡng nên đặt càng nhỏ). In kết quả. Vẽ biểu đồ.

```python
frequent_item = apriori(df, min_support=0.005, use_colnames=True)
frequent_item.head()
```

```python
association_rules(frequent_item, metric="confidence", min_threshold=0.1)
```

```python
rules = association_rules(frequent_item, metric="lift", min_threshold=1)
rules
```

```python
support=rules['support'].values
confidence=rules['confidence'].values
lift = rules['lift'].values
```

```python
plt.figure(figsize=(8,5))
plt.scatter(support, confidence, s= lift*100,alpha=0.8, c = lift)
plt.xlabel('support')
plt.ylabel('confidence')
plt.show()
```

#### Nhận xét:
Dựa trên biều đổ có thể thấy những item có khả năng được mua cùng nhau (lift>1) chỉ có 3 nhóm item. 
trong đó (frankfurter)-(other vegetables) có khoảng chênh lệch về confidence lớn khi đổi vị trí antecedents cho nhau. 2 cặp item còn lại chênh lệch confidence tương tự nhau ~0.03

### Tìm kiếm thông tin từ kết quả: trong thông tin kết quả có 'sausage' không? Nếu có thì 'sausage' kết hợp với item nào?

```python
list_combine = []
for row in rules.iterrows():
    if 'sausage'in row[1][0]:
        list_combine.append(str(row[1][1]))
print('Có thể bạn cũng muốn mua kèm với:', ' hoặc '.join(pd.Series(list_combine).map(lambda x: x[12:-3])))
```

### Cho biết 15 sản phẩm được mua nhiều nhất. Vẽ biểu đồ

```python
result = df.apply(pd.value_counts).fillna(0)
result
```

```python
df_transaction = result.iloc[1,:].sort_values(ascending=False)
df_top = df_transaction[:15]
```

```python
plt.figure(figsize=(8,4))
plt.bar(df_top.index, df_top.values, align='center')
plt.xticks(rotation='vertical')
plt.ylabel('Buying Times')
plt.xlabel('Names')
plt.title('15 best seller');
```

##### Nhận xét: 
Top 15 sản phẩm bán chạy nhất được thống kê với lưu lượng mua bán được thống kê trên 500 lượt mua.

### Cho biết 15 sản phẩm được mua ít nhất. Vẽ biểu đồ

```python
df_bottom = df_transaction[-15:]
```

```python
plt.figure(figsize=(8,4))
plt.bar(df_bottom.index, df_bottom.values, align='center')
plt.xticks(rotation='vertical')
plt.ylabel('Buying Times')
plt.xlabel('Names')
plt.title('15 worst seller');
```

##### Nhận xét:
15 sản phẩm bán ế ẩm nhất chỉ có xấp xỉ dưới 10 lượt mua.
